# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize = (5,3), dpi = 300)
# size of graph

plt.plot(x,y, label = '2x', color = 'red', linewidth = 4, marker = '.', markersize = 10, markeredgecolor = 'black',
        linestyle =':')
# label, labels the line so that we have smth to show sa legend
# color, changes d color
# linewidth, changes d size
# marker, adds markers to the points
# markersize, changes the size of the marker
# markeredge, adds outline
# markeredgecolor, color of outline
# linestyle, style ng line

# shorthand notation: fmt = [color][marker][line]
# example is: plt.plot(x,y, 'r.-', label = '2x')

## line 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, label = 'x^2') #ends at 4, does not include 5
#plots first 5 values
plt.plot(x2[4:],x2[4:]**2,'r--')
#plots the last 4 values

plt.title("eme", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
#gives title
#fontdict = puts font
#fontsize = puts size

plt.xlabel('x-axis')
plt.ylabel('y-axis')
#gives label to x and y axis

plt.xticks([0,1,2,3,4])
#adjusts x ticks
#plt.yticks([0,2,4,6,8,10])
#adjusts y ticks

plt.legend()

#plt.savefig('mygraph.png', dpi =300)

plt.show()
# removes the line thingy
```

```python
###BAR CHARTS

labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')
#shortened = 
patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    #pops the first item off when done/applied

plt.figure(figsize = (6,4))

plt.show()
```

```python
import requests
import matplotlib.pyplot as plt
import time
```

```python
api_url = 'http://api.open-notify.org/iss-now.json'
data_points = 100

def track_iss():
    latitudes, longitudes, timestamps = [], [], []
    
    for _ in range(data_points):
        response = requests.get(api_url)
        data = response.json()
        
        latitude = float(data["iss_position"]["latitude"])
        longitude = float(data['iss_position']['latitude'])
        timestamp = int(data['timestamp'])
        
        latitudes.append(latitude)
        longitudes.append(longitude)
        timestamps.append(timestamp)
        
        time.sleep(10)
        
    return latitudes, longitudes, timestamps    
```

```python
def plot_iss(latitudes, longitudes):
    plt.figure(figsize = (10,6))
    plt.plot(longitudes, latitudes, 'b.-')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('ISS Path')
    plt.grid(True)
    plt.show()
```

```python
def iss_path():
    latitudes, longitudes, _ = track_iss()
    plot_iss(latitudes, longitudes)
    
iss_path()
```
